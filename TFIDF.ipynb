```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://vijayboyapati.medium.com/the-bullish-case-for-bitcoin-6ecc8bdecc1'
reqs = requests.get(url)
```

```python
reqs.text
```

```python
soup = BeautifulSoup(requests.get(url).text, 'lxml')
```

```python
 
para = []
for data in soup.find_all("p"):
    print(data.get_text())
    para.append(data.get_text())
```

```python
print(para)
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words("english")
```

```python
from nltk.tokenize import word_tokenize
```

```python
stopwords
```

```python
stringList = ' '.join([str(item) for item in para])
```

```python
print(stringList)
```

```python
tokenized_text = word_tokenize(stringList)
```

```python
filtered_para = [word for word in tokenized_text if not word in stopwords]
```

```python
print(filtered_para)
```

```python
stringList1 = ' '.join([str(item) for item in filtered_para])
```

```python
stringList1 = [stringList1]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer, TfidfTransformer
```

```python
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(stringList1)

```

```python
print(max(vectors))
```

```python
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
```

```python
df = pd.DataFrame(denselist, columns = feature_names)
```

```python
df
```

```python
rslt_df = df.sort_values(by = 0, axis = 1, ascending = False)
```
